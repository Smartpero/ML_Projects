# Import Libraries

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression, LinearRegression, LogisticRegression
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import xgboost as xgb #Xgboost
import catboost as cat # catboost
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split, KFold, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor # decision trees
```

# A. Regression Model Implementation

## Load Dataset

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load dataset
import zipfile
zip_path = "forest+fires.zip"

with zipfile.ZipFile(zip_path) as z:
    # get first file name inside zip
    filename = z.namelist()[0]
    with z.open(filename) as f:
        df = pd.read_csv(f)

print(df)
```

```python
# display first few rows
df.head()
```

The aim of the regression problem is to predict the burned area of forest fires in the northern region of Portugal by using meteorological and other data

## Exploratory Data Analysis

```python
# Summary statistics
df.describe()
```

```python
# drop non-numeric varaible columns
data = df.drop(columns=["month", 'day'], axis=1)
```

```python
# Show correlation matrix
data.corr()
```

```python
# Visualise features vs target variable plot
features = ['DMC', 'temp', 'RH', 'wind', 'rain']
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(20, 5))
for i, feature in enumerate(features):
    sns.scatterplot(x=data[feature], y=data['area'], ax=axes[i])
    axes[i].set_title(f'{feature} vs Area')
plt.tight_layout()
plt.show()
```

## Data Preprocessing

```python
X_data = data[["DMC", "temp", "RH", "wind", "rain"]]
Y_data = data["area"]
```

```python
#check for nulls
X_data.isnull().sum()
```

```python
# Scale the target variable
scaler = StandardScaler()
Y_scaled = scaler.fit_transform(Y_data.values.reshape(-1, 1))
```

```python
Y_data_scaled = Y_scaled
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data_scaled, test_size=0.2, random_state=42)
```

## Model Training : Non-linear Regression Models

```python
# Kfold cross validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42) # kfold validation with 5 splits
```

**Xgboost Regression (+ K fold Validation)**

```python

# define grid parameters
param_grid = {
    'n_estimators': [100, 200, 300], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [None, 5, 10, 15, 20],
    "learning_rate": [0.01, 0.1, 0.2, 0.5, 1],
}

# Implement Gridsearch on the XgboostRegressor
grid_xgb =  GridSearchCV(estimator=xgb.XGBRegressor(), param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_xgb_result = grid_xgb.fit(X_train, y_train)

print (f"Best Xgboost Parameters: {grid_xgb_result.best_params_}")
```

**Decision Tree (+ K fold Validation)**

```python
# define grid parameters
param_grid = {
    'max_depth': [None, 5, 10, 15, 20],
    "min_samples_leaf": [1, 2, 4],
    "min_samples_split": [2, 5, 10],
    "max_features": ["auto", "sqrt", "log2"],
}

# Implement Gridsearch on the DecisionTreeRegressor
grid_dt = GridSearchCV(estimator=DecisionTreeRegressor(), param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_dt_result = grid_dt.fit(X_train, y_train)

print(f"Best DecsionTree Parameters: {grid_dt_result.best_params_}")
```

**Catboost Regressor (+ K fold Validation)**

```python
# define grid parameters
"""
param_grid = {
    'n_estimators': [100, 200, 300], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [None, 5, 10, 15, 20],
    "learning_rate": [0.01, 0.1, 0.2, 0.5, 1],
}

# Implement Gridsearch on the CatboostRegressor
grid_cat = GridSearchCV(estimator=cat.CatBoostRegressor(), param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_cat_result = grid_cat.fit(X_train, y_train)

print(f"Best Catboost Parameters: {grid_cat_result.best_params_}")
"""
```

GridSearchCV for the ***Catboostregressor*** takes too long to run and so we proceed to use a faster *RandomizedSearchCV* and introduce early_stopping_rounds to enhance performance

```python
# define grid parameters
param_grid = {
    'n_estimators': [100, 200, 300], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [None, 5, 10, 15, 20],
    "learning_rate": [0.01, 0.1, 0.2, 0.5],
}

# Implement RandomsearchCV on the CatboostRegressor
grid_cat = RandomizedSearchCV(estimator=cat.CatBoostRegressor(verbose=0, random_state=42, early_stopping_rounds=50), param_distributions=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_cat_result = grid_cat.fit(X_train, y_train)

print(f"Best Catboost Parameters: {grid_cat_result.best_params_}")
```

**Random Forest Regressor (+ K fold Validation)**

```python
# define grid parameters
param_grid = {
    'n_estimators': [100, 200, 300], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [None, 5, 10, 15, 20],
    "min_samples_split": [2, 5, 10],
    "max_features": ["auto", "sqrt", "log2"],
}

# Implement Gridsearch on the Randomforest
grid_rf =  GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_rf_result = grid_rf.fit(X_train, y_train)

print(f"Best Random_Forest Parameters: {grid_rf_result.best_params_}")
```

# Compare Model Performance

```python
# Evaluate Model Performance
def evaluate_model(model, X_test, y_test, name="Model"):
    y_pred = model.best_estimator_.predict(X_test) # predict with the best model estimator
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return {"Model": name, "MSE": mse, "R2": r2}

model_results = []
```

```python
# Xgboost
model_results.append(evaluate_model(grid_xgb_result, X_test, y_test, "Xgb Regressor"))

# Decision Tree regressor
model_results.append(evaluate_model(grid_dt_result, X_test, y_test, "Decision Tree Regressor"))

# Catboost Regressor
model_results.append(evaluate_model(grid_cat_result, X_test, y_test, "Catboost Regressor"))

# Random Forest Regressor
model_results.append(evaluate_model(grid_rf_result, X_test, y_test, "Random Forest"))


model_results_df = pd.DataFrame(model_results)
print(model_results_df)
```

```python
models = ["Xgboost", "Decision Tree", "Catboost", "Random Forest"]
mse_scores = model_results_df["MSE"]

plt.figure(figsize=(8, 5))
plt.bar(models, mse_scores, color=['blue', 'green', 'red', 'orange'])
plt.xlabel("Model")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Model Performance (MSE) Comparison")
plt.show()
```

## Load Dataset

The PIRvision dataset contains occupancy detection data collected from a Synchronized Low-Energy Electronically-chopped Passive Infra-Red sensing node in residential and office environments. Each observation represents 4 seconds of recorded human activity within the sensor Field-of-View (FoV).

```python
# Load dataset
import zipfile
zip_path = "pirvision_fog_presence_detection.zip"

with zipfile.ZipFile(zip_path) as z:
    # get first file name inside zip
    filename = z.namelist()[0]
    with z.open(filename) as f:
        df = pd.read_csv(f)

df.head()
```

Interestingly Target Variable "Label" is already one-hot encoded to indicate presence or absence of occupancy

## Exploratory Data Analysis

```python
# shape
df.shape
```

```python
# Summary statistics
df.describe()
```

```python
data = df.drop(columns=["Date", "Time"], axis=1)
```

```python
# Visualise features
features = ['PIR_11', 'PIR_22', 'PIR_33', 'PIR_44', 'PIR_55', "Label" ]#pick random relevant features alongside target "Label"
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(20, 5))
for i, feature in enumerate(features):
    sns.histplot(x=data[feature], ax=axes[i])
    axes[i].set_title(f'{feature}')
plt.tight_layout()
plt.show()
```

## Data Preprocessing

```python
X_data = data[['PIR_11', 'PIR_22', 'PIR_33', 'PIR_44', 'PIR_55']]
Y_data = data["Label"]
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42, stratify=Y_data)
```

## Feature Engineering

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Model Training & Performance Evaluation

```python
# Kfold cross validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42) # kfold validation with 5 splits
```

**Xgboost Classifier**

```python
# define the classifier as clf
clf_xgb = xgb.XGBClassifier()

# Map the label '3' to '2' in the target variable
y_train = y_train.replace(3, 2)
y_test = y_test.replace(3, 2)

# define grid parameters
param_grid = {
    'n_estimators': [100, 200, 300], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [None, 5, 10, 15, 20],
    "learning_rate": [0.01, 0.1, 0.2, 0.5, 1],
}

# Implement Gridsearch on the XgboostClassifier
grid_xgb =  GridSearchCV(estimator=clf_xgb, param_grid=param_grid, cv=kfold, scoring='accuracy', verbose=1, n_jobs=-1)
grid_xgb_result = grid_xgb.fit(X_train, y_train)

print (f"Best Xgboost Parameters: {grid_xgb_result.best_params_}")
```

**Decision Tree Classifier**

```python
# define the classifier as clf
clf_dt = DecisionTreeClassifier()

# define grid parameters
param_grid = {
    'max_depth': [None, 5, 10, 15, 20],
    "min_samples_leaf": [1, 2, 4],
    "min_samples_split": [2, 5, 10],
    "max_features": ["auto", "sqrt", "log2"],
}

# Implement Gridsearch on the DecisionTreeClassifier
grid_dt = GridSearchCV(estimator=clf_dt, param_grid=param_grid, cv=kfold, scoring='accuracy', verbose=1, n_jobs=-1)
grid_dt_result = grid_dt.fit(X_train, y_train)

print(f"Best DecsionTree Parameters: {grid_dt_result.best_params_}")
```

**Catboost Classifier**

```python
# define the classifier as clf
clf_cat = cat.CatBoostClassifier(verbose=0, random_state=42, early_stopping_rounds=50)

# define grid parameters
param_grid = {
    'n_estimators': [100, 200], # specify the number of weak learners(trees) in the ensemble method
    'max_depth': [3, 5, 7, 10], # altered for faster implementation
    "learning_rate": [0.01, 0.1, 0.2, 0.5],
}

# Implement RandomsearchCV on the CatboostRegressor
grid_cat = RandomizedSearchCV(estimator=clf_cat, param_distributions=param_grid, cv=kfold, scoring='accuracy', verbose=1, n_jobs=-1)
grid_cat_result = grid_cat.fit(X_train, y_train)

print(f"Best Catboost Parameters: {grid_cat_result.best_params_}")
```

**Random Forest Classifier**

```python
# define the classifier as clf
clf_rf = RandomForestClassifier()

# define grid parameters
param_grid = {
    'n_estimators': [50, 100, 200], # altered for faster implementation
    'max_depth': [5, 10, 15, 20],
    "min_samples_split": [2, 5],
    "max_features": ["sqrt", "log2"], # altered for faster implementation
}

# Implement RandomizedSearchCV (instead of GridSearchCV) on the Randomforest
grid_rf =  RandomizedSearchCV(estimator=clf_rf, param_distributions=param_grid, cv=kfold, scoring='accuracy', verbose=1, n_jobs=-1)
grid_rf_result = grid_rf.fit(X_train, y_train)

print(f"Best Random_Forest Parameters: {grid_rf_result.best_params_}")

```

# Compare Model Performance

```python
def eval_model(model, X_test, y_test, name="Model"):
    y_pred = model.best_estimator_.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    return {"Model": name, "Accuracy": accuracy, "Classification Report": classification_rep, "Confusion Matrix" : conf_matrix}

model_results = []
```

```python
# Xgboost Classifier
model_results.append(eval_model(grid_xgb_result, X_test, y_test, "Xgboost Classifier"))

# Decision Tree Classifier
model_results.append(eval_model(grid_dt_result, X_test, y_test, "Decision Tree Classifier"))

# Catboost Classifier
model_results.append(eval_model(grid_cat_result, X_test, y_test, "Catboost Classifier"))

# Random Forest Classifier
model_results.append(eval_model(grid_rf_result, X_test, y_test, "Random Forest Classifier"))


model_results_df = pd.DataFrame(model_results)
print(model_results_df)
```

### Visualise Model Confusion Matrices

```python
# Confusion matrix from the model_results_df for the Xgboost Classifier
conf_matrix = model_results_df.loc[0, 'Confusion Matrix']
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Xgboost Classifier')
plt.show()

# Confusion matrix from the model_results_df for Decision Tree Classifier
conf_matrix = model_results_df.loc[1, 'Confusion Matrix']
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.show()

# Confusion matrix from the model_results_df for Catboost Classifier
conf_matrix = model_results_df.loc[2, 'Confusion Matrix']
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='inferno')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Catboost Classifier')
plt.show()

# Confusion matrix from the model_results_df for Random Forest Classifier
conf_matrix = model_results_df.loc[3, 'Confusion Matrix']
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='magma')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Random Forest Classifier')
plt.show()
```
