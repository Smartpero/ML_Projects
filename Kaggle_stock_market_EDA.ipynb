# HTML Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import re  # for number extraction

# url
url = "https://www.kaggle.com/datasets/kshitijsaini121/stock-market-prediction-for-july-2025-dataset"


response = requests.get(url)

# Parse HTML

soup =  BeautifulSoup(response.text, "html.parser")

title = soup.title.text

paragraphs = soup.find_all("p")

print(title)
all_numbers = []
for p in paragraphs:
    numbers = re.findall(r'\d+', p.text)
    all_numbers.extend(numbers)

```

HTML scraping cannot be used for stock market data hence we can only print the text in the data

# Exploratory Data Analysis

```python
#Summary statistics
df.describe()
```

```python
# display data
data = df.loc[:, "Date" : 'Close Price']
data.dropna(inplace=True)
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
plt.hist(data['Close Price'], bins=20)
plt.title("Distribution of Daily Close Prices")
```

```python
#daily returns
returns = data['Close Price'].pct_change()
returns.dropna(inplace=True)
returns.hist(bins=10)
plt.title("Distribution of Daily Returns")
```
